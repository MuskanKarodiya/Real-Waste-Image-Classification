```python
%%writefile app.py
import streamlit as st
import numpy as np
from PIL import Image
from tensorflow.keras.models import load_model  # вњ… Load your custom model

# вњ… Load model from path (make sure it's your latest trained model)
model = load_model("/content/drive/MyDrive/Colab Notebooks/waste_classifier.h5")

# вњ… Class labels
class_names = [
    'Cardboard', 'Food Organics', 'Glass', 'Metal', 'Miscellaneous Trash',
    'Paper', 'Plastic', 'Textile Trash', 'Vegetation'
]

# вњ… Streamlit UI
st.title("в™»пёЏ Real Waste Image Classification")
st.write("Upload an image of waste to classify it into the correct category.")

uploaded_file = st.file_uploader("Choose a waste image...", type=["jpg", "jpeg", "png", "jfif"])

if uploaded_file is not None:
    image = Image.open(uploaded_file).convert('RGB')
    st.image(image, caption="Uploaded Image", use_container_width=True)

    image = image.resize((160, 160))
    img_array = np.array(image) / 255.0  # вњ… Match your model's normalization
    img_array = np.expand_dims(img_array, axis=0)

    prediction = model.predict(img_array)
    predicted_class = class_names[np.argmax(prediction)]
    confidence = np.max(prediction) * 100

    st.markdown(f"### рџ§  Prediction: **{predicted_class}**")
    st.markdown(f"### рџ”Ќ Confidence: **{confidence:.2f}%**")
```
